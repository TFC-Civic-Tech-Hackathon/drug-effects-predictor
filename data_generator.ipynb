```python
from faker import Faker
import random

fake = Faker()

def generate_patient_record():
    name = fake.name()
    age = random.randint(4, 0)
    if age < 18:
        height = random.randint(100, 150)  # Height between 100-150 cm for age below 18
        weight = random.randint(30, 50)      # Weight less than 50 kg for age below 18
    else:
        height = random.randint(150, 200)  # Height between 150-200 cm for age 18 and above
        weight = random.randint(40, 120)   # in kilograms
    
    gender = random.choice(['Male', 'Female'])
    
    if gender == 'Male':
        pregnancy = 'No'
    else:
        pregnancy = random.choice(['Yes', 'No'])
    
    past_medical_conditions = random.choice(["a", "b", "c", "d", "e"])  # Generating a random sentence for past medical conditions
    return {
        'Age': age,
        'Gender': gender,
        'Height(cm)': height,
        'Weight(kg)': weight,
        'Pregnancy': pregnancy,
        'Past Medical Conditions': past_medical_conditions
    }

def generate_patient_records(num_records=10):
    records = []
    for _ in range(num_records):
        records.append(generate_patient_record())
    return records

# Generating 10 patient records
patient_records = generate_patient_records(1000)

# Displaying the records
for idx, record in enumerate(patient_records, 1):
    print(f"Patient {idx}: {record}")
```
